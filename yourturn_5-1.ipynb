```python
import requests

url = 'https://opendata.cwb.gov.tw/api/v1/rest/datastore/O-A0001-001?Authorization='
r = requests.get(url)

print(r)
print(r.status_code)
```

```python
print(r.text)
```

```python
import json

j = json.loads(r.text)
print(j)
```

```python
j
```

```python
locs = j['records']['location']
locs
```

```python
print(f'{len(locs)} sites')
```

```python
for loc in locs:
    if loc['locationName'] == '福山':
        print(loc)
```
